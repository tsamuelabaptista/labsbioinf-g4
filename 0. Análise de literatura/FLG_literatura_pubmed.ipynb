```python
from Bio import Entrez
```

```python
Entrez.email = "...@gmail.com" 

def search_pubmed(query, max_results=5):
    handle = Entrez.esearch(db="pubmed", term=query, retmax=max_results)
    record = Entrez.read(handle)
    handle.close()
    return record["IdList"]

def fetch_pubmed_articles(article_ids):
    handle = Entrez.efetch(db="pubmed", id=article_ids, rettype="medline", retmode="text")
    records = handle.read()
    handle.close()
    return records
```

```python
search_query = "flg human"

article_ids = search_pubmed(search_query)

if article_ids:
        pubmed_records = fetch_pubmed_articles(article_ids)
        print(pubmed_records)
```
